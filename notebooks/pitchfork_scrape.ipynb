#### Setup

```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
import json
import random
```

```python
headers = {'User-Agent': f'ds6021-project/0.0 (jhb7ek@virginia.edu) python-requests/{requests.__version__}'}
headers
```

```python
prefix = 'https://www.pitchfork.com'
page_url_prefix = 'https://pitchfork.com/reviews/albums/?page='
url_list = []
```

#### Define function to scrape album review links off of each page

```python
def get_album_urls(url):
    r = requests.get(url, headers=headers)
    mysoup = BeautifulSoup(r.text, 'html.parser')
    album_urls = mysoup.find_all('a', 'summary-item__hed-link')
    return [prefix + review_link['href'] for review_link in album_urls]
```

```python
for i in range(1, 94):
    url = page_url_prefix + str(i)
    urls = get_album_urls(url)
    url_list.extend(urls)
    print(f'Page {i} completed, retrieved {len(urls)} links, total {len(url_list)}')

    # Save progress as it runs
    with open("album_urls.json", "w") as f:
        json.dump(url_list, f, indent=2)
    
    time.sleep(random.uniform(1, 3))
```
