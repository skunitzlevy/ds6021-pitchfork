```python
import pandas as pd

# Load the CSV file containing artist names
artists_df = pd.read_csv('out/pitchfork_reviews.csv')
artist_names = artists_df['artist'].unique().tolist()
print(f"Found {len(artist_names)} unique artists.")
```

```python
# Save the unique artist names to a new CSV file
pd.DataFrame(artist_names, columns=['artist']).to_csv('spotify/unique_artists.csv', index=False)
print("Unique artist names saved to 'out/unique_artists.csv'.")
```
