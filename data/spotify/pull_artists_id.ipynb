```python
from spotify_functions import *
import pandas as pd

# Load the CSV file containing unique artist names
artists_df = pd.read_csv("unique_artists.csv")
```

```python
# Get Spotify artist IDs for each artist name
artist_ids = []
for artist in artists_df['artist']:
    artist_id = get_artist_id(artist, access_token)
    artist_ids.append({'artist': artist, 'artist_id': artist_id})
    print(f"Artist: {artist}, Spotify ID: {artist_id}")
```

```python
# Save the artist names and their corresponding Spotify IDs to a new CSV file
artist_ids_df = pd.DataFrame(artist_ids)
artist_ids_df.to_csv('artist_ids.csv', index=False)
print("Artist names and Spotify IDs saved to 'spotify/artist_ids.csv'.")
```
